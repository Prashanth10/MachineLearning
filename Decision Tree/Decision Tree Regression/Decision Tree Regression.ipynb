```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
np.random.seed(42)
m = 250
X = np.random.rand(m,1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m,1)/10
```

```python
plt.figure(figsize=(10,5))
plt.title("ScatterPlot")
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
dt_tree_reg2 = DecisionTreeRegressor(max_depth=3, random_state=42)
```

```python
dt_tree_reg.fit(X,y)
```

```python
dt_tree_reg2.fit(X,y)
```

```python
plt.figure(figsize=(12,4))

X_temp = X.copy()
X_temp.sort(axis=0)
# X_temp
y_predict = dt_tree_reg.predict(X_temp)
y_predict2 = dt_tree_reg2.predict(X_temp)

plt.subplot(121)
plt.scatter(X, y, marker='o')
plt.plot(X_temp, dt_tree_reg.predict(X_temp), "r.-", linewidth=2)
plt.axis([0,1,-0.2,1.1])
plt.xlabel("X", fontsize=14)
plt.ylabel("$y$", fontsize=14, rotation=0)
plt.title("maxdepth = 2", fontsize=16)

plt.subplot(122)
plt.scatter(X,y,marker='o')
plt.plot(X_temp, dt_tree_reg2.predict(X_temp), "r.-",linewidth=2)
plt.axis([0,1,-0.2,1.1])
plt.xlabel("X", fontsize=14)
plt.title("maxdepth = 3", fontsize=16)
```

```python
tree_reg1 = DecisionTreeRegressor(random_state=42)
tree_reg2 = DecisionTreeRegressor(random_state=42, min_samples_leaf=10)
tree_reg1.fit(X, y)
tree_reg2.fit(X, y)

x1 = np.linspace(0, 1, 500).reshape(-1, 1)
y_pred1 = tree_reg1.predict(x1)
y_pred2 = tree_reg2.predict(x1)

plt.figure(figsize=(11, 4))

plt.subplot(121)
plt.scatter(X, y, marker='o')
plt.plot(x1, y_pred1, "r.-", linewidth=2)
plt.axis([0, 1, -0.2, 1.1])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", fontsize=18, rotation=0)
plt.legend(loc="upper center", fontsize=18)
plt.title("No restrictions", fontsize=14)

plt.subplot(122)
plt.scatter(X, y, marker="o")
plt.plot(x1, y_pred2, "r.-", linewidth=2)
plt.axis([0, 1, -0.2, 1.1])
plt.xlabel("$x_1$", fontsize=18)
plt.title("min_samples_leaf={}".format(tree_reg2.min_samples_leaf), fontsize=14)

plt.show()
```
